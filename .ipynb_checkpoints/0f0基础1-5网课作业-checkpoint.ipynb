4.2 投一个骰子，1\6概率为6，投了6次，重复60轮，平均每轮骰子为6的次数时多少？

```python
import matplotlib.pyplot as plt
from random import randint
```

```python
class Die():
    def __init__(self,num_side=6):
        self.num_side=num_side
    def roll(self):
        return randint(1,self.num_side)
die = Die()
total = []
for term in range(60):
    s = 0
    for i in range(6):
        if die.roll() == 6:
            s = s+1
    total.append(s)
total = np.array(total)
plt.figure(figsize=(15, 7))
plt.hist(total,[-0.5,0.5,1.5,2.5,3.5,4.5,5.5,6.5])
plt.xlabel('Value')
plt.ylabel('Occurences')
plt.legend(['BinomiaRandomVariable'])
plt.show()
```

5.2 用signaldigger分析沪深300成分股里pe因子在5日、30日、60日持有期下的因子表现。
（要求：a.过滤掉非沪深300成分股。b.定义可买入卖出的条件-未停牌、未涨跌停 c.用开盘价计算持有期收益 d.调用因子指标分析和可视化分析的方法，查看效果)

```python
dv = DataView()
dv.load_dataview(dataview_folder)
```

```python
def mask_index_member():
    df_index_member = dv.get_ts('index_member')
    mask_index_member = df_index_member == 0
    return mask_index_member

# 定义可买卖条件——未停牌、未涨跌停
def limit_up_down():
    trade_status = dv.get_ts('trade_status')
    mask_sus = trade_status == 0
    # 涨停
    dv.add_formula('up_limit', '(close - Delay(close, 1)) / Delay(close, 1) > 0.095', is_quarterly=False, add_data=True)
    # 跌停
    dv.add_formula('down_limit', '(close - Delay(close, 1)) / Delay(close, 1) < -0.095', is_quarterly=False, add_data=True)
    can_enter = np.logical_and(dv.get_ts('up_limit') < 1, ~mask_sus) 
    can_exit = np.logical_and(dv.get_ts('down_limit') < 1, ~mask_sus) 
    return can_enter,can_exit

mask = mask_index_member()
can_enter,can_exit = limit_up_down()
```

```python
from jaqs_fxdayu.research import SignalDigger
obj1 = SignalDigger(output_folder='./output',
                   output_format='pdf')
obj2 = SignalDigger(output_folder='./output',
                   output_format='pdf')
obj3 = SignalDigger(output_folder='./output',
                   output_format='pdf')
obj1.process_signal_before_analysis(signal=dv.get_ts("pe"),
                                   price=dv.get_ts("open_adj"),
                                   high=dv.get_ts("high_adj"), 
                                   low=dv.get_ts("low_adj"),
                                   group=dv.get_ts("sw1"),
                                   n_quantiles=5,
                                   mask=mask,
                                   can_enter=can_enter,
                                   can_exit=can_exit,
                                   period=5,
                                   benchmark_price=dv.data_benchmark,
                                   commission = 0.0008,
                                   )
obj2.process_signal_before_analysis(signal=dv.get_ts("pe"),
                                   price=dv.get_ts("open_adj"),
                                   high=dv.get_ts("high_adj"), 
                                   low=dv.get_ts("low_adj"),
                                   group=dv.get_ts("sw1"),
                                   n_quantiles=5,
                                   mask=mask,
                                   can_enter=can_enter,
                                   can_exit=can_exit,
                                   period=30,
                                   benchmark_price=dv.data_benchmark,
                                   commission = 0.0008,
                                   )
obj3.process_signal_before_analysis(signal=dv.get_ts("pe"),
                                   price=dv.get_ts("open_adj"),
                                   high=dv.get_ts("high_adj"), 
                                   low=dv.get_ts("low_adj"),
                                   group=dv.get_ts("sw1"),
                                   n_quantiles=5,
                                   mask=mask,
                                   can_enter=can_enter,
                                   can_exit=can_exit,
                                   period=60,
                                   benchmark_price=dv.data_benchmark,
                                   commission = 0.0008,
                                   )
obj1.create_full_report()
obj2.create_full_report()
obj3.create_full_report()
plt.show()
```

```python
from jaqs_fxdayu.research.signaldigger.analysis import analysis
signal_data1 = obj1.signal_data
signal_data2 = obj2.signal_data
signal_data3 = obj3.signal_data
result1 = analysis(signal_data1, is_event=False, period=5)
result2 = analysis(signal_data2, is_event=False, period=30)
result3 = analysis(signal_data3, is_event=False, period=60)
print('period=5')
print("——ic分析——")
print(result1["ic"])
print("——选股收益分析——")
print(result1["ret"])
print("——最大潜在盈利/亏损分析——")
print(result1["space"])
print('period=30')
print("——ic分析——")
print(result2["ic"])
print("——选股收益分析——")
print(result2["ret"])
print("——最大潜在盈利/亏损分析——")
print(result2["space"])
print('period=60')
print("——ic分析——")
print(result3["ic"])
print("——选股收益分析——")
print(result3["ret"])
print("——最大潜在盈利/亏损分析——")
print(result3["space"])
```

5.4 用append_df方法定义CCI （提示：a. 需处理好停牌期的K线数据 b.可以用内置好的signal_function_mod.ta方法调用talib库计算CCI）

```python
from jaqs_fxdayu.research.signaldigger import process
High = dv.get_ts("high_adj")
Low = dv.get_ts("low_adj")
Close = dv.get_ts("close_adj")
trade_status = dv.get_ts('trade_status')
mask_sus = trade_status == 0
# 剔除掉停牌期的数据　再计算指标
high_masked = process._mask_df(High,mask=mask_sus)
low_masked = process._mask_df(Low,mask=mask_sus)
close_masked = process._mask_df(Close,mask=mask_sus)
```

```python
from jaqs_fxdayu.data import signal_function_mod as sfm
CCI = sfm.ta(ta_method='CCI',
             ta_column=0, 
             High=high_masked, 
             Low=low_masked, 
             Close=close_masked,
             Volume=None,
             timeperiod=10)
```

```python
dv.append_df(CCI,'CCI')
```
